```python
#2
import pandas as pd

Speed= pd.read_csv('2281305-1.csv')
S=Speed.loc[:,['DATE','SPEED']]
#引入了文件，并提前在excel表格中的最后一列进行分列，将风速列命名为SPEED。
#选择出时间和风速的信息
S['DATE'] = pd.to_datetime(S['DATE'])
#借鉴此处的日期转换方法https://blog.csdn.net/mhywoniu/article/details/78514664
S
# for i in S['2010-1':'2020-1']：#没能完成提取10年间的各个月份

#     m=S['i'].mean()#对每个月的风速求平均值
#     a = pd.Series({'i':m})#添加到一个系列中
    
# S.plot(kind='line')

M=S.groupby([S['DATE'].dt.year,S['DATE'].dt.month])['SPEED'].mean()
#一开始考虑使用for循环将每月的结果列入一个新表格后作图。经左小幸指导后采用分组的方式完成。
M.plot(kind='line')
```
