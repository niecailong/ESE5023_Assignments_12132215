```python
#1
import pandas as pd

Sig_Eqs= pd.read_csv('earthquakes-2021-10-23_11-03-55_+0800.txt',sep='\t')
#建立函数，实现统计每年地震的次数和找到每年发生最大震级发生的日期
def CountEq_LargestEq(x) :
    
    b=Sig_Eqs.loc[Sig_Eqs['Country'] ==x] .sort_values('Mag',ascending=False).head(1)['Year'].values[0]

 
    return(b)

def c(y) :
    a=Sig_Eqs[Sig_Eqs['Year']>-2150].groupby('Country').count()['Year']
    d=a[y]
    return(d)

#建立一个字典，先把最大震级发生的日期和国家添加进去。
#请实验室的程战文师兄指导完成的这一步,自己理解后完善，在字典中第一列为日期，第二列为地震发生的次数
b = {}
for i in Sig_Eqs.Country:
    if pd.isnull(i):
        continue
    else:
        b[i] = [CountEq_LargestEq(i),c(i)]
        print(i,CountEq_LargestEq(i),c(i))
for key,item in b:
    print(key,item)
```
